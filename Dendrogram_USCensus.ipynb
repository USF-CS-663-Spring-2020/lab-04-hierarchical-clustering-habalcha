```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

from scipy.cluster.hierarchy import linkage, dendrogram

#import the dataset
df = pd.read_csv('USCensus1990-2000rows.data.txt')
df
```

```python
#cluster df using linkage
X = linkage(df, 'ward')
```

```python
#plot the dendrogram from df after linkage
plt.figure(figsize = (25, 15))
dendrogram(X)
plt.show()
```
