```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
from scipy.cluster.hierarchy import linkage, dendrogram

df = pd.read_csv('NIPS_1987-2015.csv')
```

```python
#remove row with the word labels and keep only the frequency of each word
df = df.iloc[:, 1:]
```

```python
#cluser the data in df using linkage
X = linkage(df, 'ward')
```

```python
#plot dendrogram
plt.figure(figsize = (20, 10))
dendrogram(X)
plt.show()
```
